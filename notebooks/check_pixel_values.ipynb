```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img0 = np.array(Image.open("../data/girl2.png"))
img1 = np.array(Image.open("../data/tower1.png"))
img2 = np.array(Image.open("../data/tower2.png"))
```

```python
fig, axs = plt.subplots(1,3)
axs[0].imshow(img0)
axs[0].set_title("var: " + str(int(img0.var())))
axs[1].imshow(img1)
axs[1].set_title("var: " + str(int(img1.var())))
axs[2].imshow(img2)
axs[2].set_title("var: " + str(int(img2.var())))
```
